```python
plt.figure(figsize=(15,15))

sns.heatmap(data= df.isnull())
```

```python
null_features = []

for x in df:
    if df[x].isnull().sum() > 0:
        null_features.append(x)
print(null_features)
```

```python
empty_features = [x for x in df if df[x].isnull().sum() >= 0 and df[x].dtypes!='O' ]
#print(empty_features)
for y in empty_features:
     df[y] = df[y].fillna( df[y].mean())
```

```python
plt.figure(figsize=(15,15))

sns.heatmap(data= df.isnull())
```

```python
empty_features = [x for x in df if df[x].isnull().sum() > 0 and df[x].dtypes=='O' ]
#print(empty_features)
for y in empty_features:
     df[y] = df[y].fillna( df[y].mode()[0] )
```

```python
plt.figure(figsize=(15,15))

sns.heatmap(data= df.isnull())
```
