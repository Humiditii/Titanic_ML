```python
class FeatureEngr():
    def __init__ (self, data):
        self.data = data;
    def plot_empty(self):
        plt.figure(figsize=(20,20))
        return sns.heatmap(self.data.isnull())
    def drop_clean(self):
        a = list()
        b = list()
        for x in self.data:
            count_null = self.data[x].isnull().sum();
            percentage_null = (count_null/len(self.data) )* 100
            if percentage_null > 70:
                self.data.drop([x], 1, inplace=True)
            elif self.data[x].isnull().sum() > 0:
                if self.data[x].dtypes == 'O':
                      self.data[x] = self.data[x].fillna(self.data[x].mode()[0])
                elif self.data[x].dtypes != 'O':
                    self.data[x] = self.data[x].fillna(self.data[x].mean())
    
        return self.data
    def transformer_oh(self, features):
        self.data = pd.get_dummies(self.data, prefix_sep='_', columns=features)
        return self.data
    def transformer_le(self, features, encoder):
        for col in features:
            self.data[col] = encoder.fit_transform(self.data[col])
        return self.data
        
                    
                
#                 if (self.data[x].dtypes == 'O' and self.data[x].isnull().sum() > 0):
#                     self.data[x] = object_imputer.fit_transform(self.data)
#                 elif (self.data[x].dtypes != 'O' and self.data[x].isnull().sum() > 0):
#                     self.data[x] = int_imputer.fit_transform(self.data)
#                 else:
#                     self.data
#         return self.data
                    
                    
```

```python
def get_esemble():
    models = list()
    models.append(('gb', gb))
    models.append(('rf', rf))
    models.append(('xgb', xgbC))
    
    esemble = VotingClassifier(estimators=models, voting='hard')
    return esemble
```

```python
folds = StratifiedKFold(n_splits=10, random_state=42,shuffle=True)
```

```python
folds
```

```python
class Model():
    def __init__(self, x_train, x_test, y_train, y_test):
        self.x_train = x_train
        self.x_test = x_test
        self.y_train = y_train
        self.y_test = y_test
    def ml_model(self, model):
        self.model = model
        model.fit(x_train, y_train)
        pred_train = model.predict(x_train)
        pred_test = model.predict(x_test)
        test_score = accuracy_score(y_test, pred_test) * 100 
        train_score = accuracy_score(y_train, pred_train )*100
        
        accuracy = dict()
        accuracy['Training'] = train_score
        accuracy['Testing'] = test_score
        
        return accuracy
```
