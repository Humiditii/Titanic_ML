## Let's Join the two datasets together to form a single dataset

```python
##Defining a function or even creating a python class

# with function 

def join(train, test):
    copy_train = train.copy()
    copy_test = test.copy()
    
    copy_train.drop(['Survived'], 1, inplace=True)
    
    return copy_train.append(copy_test)

# Will explain this better at the end of the class
```

```python
df = join(df_train, df_test)
```
