```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
test_data = pd.read_csv('test.csv')
train_data = pd.read_csv('train.csv')
```

```python
class TitanicPeparation():
    def __init__(self, test, train):
        self.test = test;
        self.train = train;
    def join_data(self):
        new_test = self.test.copy()
        new_train = self.train.copy();
        new_train.drop(['Survived'], 1, inplace=True )
        return new_train.append(new_test);
```

```python
titanic = TitanicPeparation(test_data, train_data)
```

```python
df = titanic.join_data()
```

```python
class FeatureEngr():
    def __init__ (self, data):
        self.data = data;
    def plot_empty(self):
        plt.figure(figsize=(20,20))
        return sns.heatmap(self.data.isnull())
    def drop_clean(self):
        a = list()
        b = list()
        for x in self.data:
            count_null = self.data[x].isnull().sum();
            percentage_null = (count_null/len(self.data) )* 100
            if percentage_null > 70:
                self.data.drop([x], 1, inplace=True)
            elif self.data[x].isnull().sum() > 0:
                if self.data[x].dtypes == 'O':
                      self.data[x] = self.data[x].fillna(self.data[x].mode()[0])
                elif self.data[x].dtypes != 'O':
                    self.data[x] = self.data[x].fillna(self.data[x].mean())
    
        return self.data
    def transformer_oh(self, features):
        self.data = pd.get_dummies(self.data, prefix_sep='_', columns=features)
        return self.data
    def transformer_le(self, features, encoder):
        for col in features:
            self.data[col] = encoder.fit_transform(self.data[col])
        return self.data
```

```python
Feature_Engr = FeatureEngr(df)
```
